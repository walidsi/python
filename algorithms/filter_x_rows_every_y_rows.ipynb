```python
# write a function to filter a dataset such that every x rows it selects the next y rows
def filter_out_x_rows_every_y_rows(arr: list, x: int, y: int):
    """Filters a dataset such that every x rows it selects the next y rows

    Args:
        arr (list): array to filter
        x (int): number of rows to skip
        y (int): number of rows to select

    Return:
        filtered_arr (list): filtered array
    """
    filtered_arr = []

    for i in range(x, len(arr), x+y):
        filtered_arr.append(arr[i:i+y])

    return filtered_arr
```

```python
arr = [i for i in range(1, 100)]

filtered_arr = filter_out_x_rows_every_y_rows(arr, 3, 1)

print(filtered_arr)
```

```python
import pandas as pd

# write a function to filter a dataset such that every x rows it selects the next y rows
def filter_out_x_rows_every_y_rows2(df: pd.DataFrame, x: int, y: int):
    """Filters a dataset such that every x rows it selects the next y rows

    Args:
        arr (list): array to filter
        x (int): number of rows to skip
        y (int): number of rows to select

    Return:
        filtered_arr (list): filtered array
    """
    out_df = pd.DataFrame()

    for i in range(x, len(df), x+y):
        out_df = out_df.append(df.iloc[i:i+y])

    return out_df
```
